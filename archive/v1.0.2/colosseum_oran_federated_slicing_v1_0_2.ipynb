```python
# @title Cell 4: è³‡æ–™è¼‰å…¥ã€é è™•ç†èˆ‡ç¸®æ”¾ (å·²ä¿®æ­£ FutureWarning)
# ===== å°å…¥å‡½å¼åº« =====
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import joblib
import warnings

# å¿½ç•¥ FutureWarning
warnings.simplefilter(action='ignore', category=FutureWarning)


# ===== è¼‰å…¥èˆ‡æª¢æŸ¥è¨­å®š =====
print("ğŸš€ é–‹å§‹è¼‰å…¥èˆ‡é è™•ç†è³‡æ–™...")
print("="*60)
# try...except æ˜¯ä¸€å€‹å¥½çš„é˜²å‘†è¨­è¨ˆï¼Œç¢ºä¿ä½¿ç”¨è€…å·²ç¶“å…ˆåŸ·è¡Œäº† Cell 1 çš„è¨­å®š
try:
    DATA_FILE_PATH
    SAMPLES_PER_CLIENT
    TRAIN_TEST_SPLIT_RATIO
    RANDOM_STATE
    ARTIFACTS_SAVE_PATH
except NameError:
    print("âŒ è«‹å…ˆåŸ·è¡Œ Cell 1 (å°ˆæ¡ˆçµ„æ…‹è¨­å®š)")
    raise

# ===== è¼‰å…¥ä¸»è³‡æ–™é›† =====
# å¾ Parquet æª”æ¡ˆè®€å–è³‡æ–™åˆ° Pandas DataFrame
# Parquet æ˜¯ä¸€ç¨®é«˜æ•ˆç‡çš„æ¬„ä½å¼å„²å­˜æ ¼å¼ï¼Œæ¯” CSV æ›´å¿«ã€æ›´çœç©ºé–“
try:
    df = pd.read_parquet(DATA_FILE_PATH)
    print(f"âœ… æˆåŠŸè¼‰å…¥è³‡æ–™: {len(df):,} ç­†è¨˜éŒ„")
    print(f"   åŸºç«™æ•¸é‡: {df['bs_id'].nunique()}")
except Exception as e:
    print(f"âŒ è¼‰å…¥è³‡æ–™å¤±æ•—: {e}")
    raise

# ===== å®šç¾©ç‰¹å¾µèˆ‡ç›®æ¨™ =====
# feature_columns: ç”¨ä¾†é æ¸¬çš„è¼¸å…¥ç‰¹å¾µæ¬„ä½åç¨±åˆ—è¡¨ (å…± 13 å€‹)
feature_columns = [
    'num_ues', 'slice_id', 'sched_policy_num', 'allocated_rbgs',
    'sum_requested_prbs', 'sum_granted_prbs', 'prb_utilization',
    'throughput_efficiency', 'qos_score', 'network_load',
    'hour', 'minute', 'day_of_week'
]
# target_column: æˆ‘å€‘æƒ³è¦é æ¸¬çš„ç›®æ¨™æ¬„ä½
target_column = 'allocation_efficiency'

# å†æ¬¡é˜²å‘†ï¼Œç¢ºèª DataFrame ä¸­åŒ…å«æ‰€æœ‰éœ€è¦çš„æ¬„ä½
missing_cols = set(feature_columns + [target_column, 'bs_id']) - set(df.columns)
if missing_cols:
    raise ValueError(f"è³‡æ–™ç¼ºå°‘å¿…è¦æ¬„ä½: {missing_cols}")

# ===== ç‚ºæ¯å€‹å®¢æˆ¶ç«¯ç¨ç«‹è¨ˆç®—ç¸®æ”¾å™¨ (Scaler) =====
# é€™æ˜¯è¯é‚¦å­¸ç¿’ä¸­ä¸€å€‹éå¸¸é‡è¦çš„æ­¥é©Ÿï¼Œå¯ä»¥å¢å¼·éš±ç§ä¸¦é©æ‡‰è³‡æ–™ç•°è³ªæ€§
client_scalers = {}  # ç”¨ä¾†å„²å­˜æ¯å€‹å®¢æˆ¶ç«¯ (bs_id) è‡ªå·±çš„ç¸®æ”¾å™¨ç‰©ä»¶
client_stats = []    # ç”¨ä¾†å„²å­˜æ¯å€‹å®¢æˆ¶ç«¯çš„çµ±è¨ˆæ•¸æ“š (å¹³å‡å€¼ã€æ¨™æº–å·®ç­‰)

print("\nğŸ“Š ç‚ºå„å®¢æˆ¶ç«¯åˆ†åˆ¥è¨ˆç®—ç¸®æ”¾åƒæ•¸...")

# éæ­·è³‡æ–™ä¸­æ‰€æœ‰ä¸é‡è¤‡çš„åŸºåœ°å° ID (bs_id)
for client_id in sorted(df['bs_id'].unique()):
    client_df = df[df['bs_id'] == client_id]
    sample_size = min(len(client_df), SAMPLES_PER_CLIENT)
    sampled_df = client_df.sample(n=sample_size, random_state=RANDOM_STATE)
    X_client = sampled_df[feature_columns]
    y_client = sampled_df[[target_column]]
    X_train, X_test, y_train, y_test = train_test_split(
        X_client, y_client, train_size=TRAIN_TEST_SPLIT_RATIO, random_state=RANDOM_STATE
    )
    feature_scaler = StandardScaler()
    target_scaler = StandardScaler()
    feature_scaler.fit(X_train)
    target_scaler.fit(y_train)
    client_scalers[client_id] = {
        'feature_scaler': feature_scaler,
        'target_scaler': target_scaler,
        'feature_mean': feature_scaler.mean_,
        'feature_std': feature_scaler.scale_,
        'target_mean': target_scaler.mean_,
        'target_std': target_scaler.scale_,
        'sample_count': len(X_train)
    }
    client_stats.append({
        'client_id': client_id,
        'sample_count': len(X_train),
        'feature_mean': feature_scaler.mean_,
        'feature_std': feature_scaler.scale_,
        'target_mean': target_scaler.mean_[0],
        'target_std': target_scaler.scale_[0]
    })
    print(f"  - Client {client_id}: {len(X_train)} è¨“ç·´æ¨£æœ¬")

# ===== è¨ˆç®—å…¨åŸŸåŠ æ¬Šå¹³å‡ç¸®æ”¾åƒæ•¸ =====
print("\nğŸ“Š è¨ˆç®—å…¨åŸŸåŠ æ¬Šå¹³å‡ç¸®æ”¾åƒæ•¸...")
total_samples = sum(stats['sample_count'] for stats in client_stats)
global_feature_mean = np.zeros(len(feature_columns))
global_feature_std = np.zeros(len(feature_columns))
global_target_mean = 0
global_target_std = 0

for stats in client_stats:
    weight = stats['sample_count'] / total_samples
    global_feature_mean += weight * stats['feature_mean']
    global_feature_std += weight * stats['feature_std']
    global_target_mean += weight * stats['target_mean']
    global_target_std += weight * stats['target_std']


# ===== å‰µå»ºä¸¦è¨­å®šå…¨åŸŸç¸®æ”¾å™¨ =====
global_feature_scaler = StandardScaler()
global_target_scaler = StandardScaler()
global_feature_scaler.mean_ = global_feature_mean
global_feature_scaler.scale_ = global_feature_std
global_feature_scaler.n_features_in_ = len(feature_columns)
try:
    global_feature_scaler.feature_names_in_ = np.array(feature_columns)
except AttributeError:
    pass
global_target_scaler.mean_ = np.array([global_target_mean])
global_target_scaler.scale_ = np.array([global_target_std])
global_target_scaler.n_features_in_ = 1


# ===== ä½¿ç”¨å„è‡ªçš„ç¸®æ”¾å™¨è½‰æ›è³‡æ–™ =====
print("\nğŸ“Š ä½¿ç”¨å„å®¢æˆ¶ç«¯ç¨ç«‹çš„ç¸®æ”¾å™¨é€²è¡Œè³‡æ–™ç¸®æ”¾...")
df_scaled = df.copy()

# å†æ¬¡éæ­·æ‰€æœ‰å®¢æˆ¶ç«¯
for client_id, scalers in client_scalers.items():
    client_mask = df_scaled['bs_id'] == client_id

    # å°‡æµ®é»æ•¸è½‰æ›å›åŸå§‹æ•¸æ“šé¡å‹
    transformed_features = scalers['feature_scaler'].transform(df_scaled.loc[client_mask, feature_columns])
    for i, col in enumerate(feature_columns):
        original_dtype = df[col].dtype
        df_scaled.loc[client_mask, col] = transformed_features[:, i].astype(original_dtype)

    transformed_target = scalers['target_scaler'].transform(df_scaled.loc[client_mask, [target_column]])
    original_target_dtype = df[target_column].dtype
    df_scaled.loc[client_mask, target_column] = transformed_target.flatten().astype(original_target_dtype)

print("âœ… å„å®¢æˆ¶ç«¯ç¨ç«‹ç¸®æ”¾å®Œæˆï¼")

# ===== ä¿å­˜ç”¢å‡ºç‰©ä»¶ =====
artifacts = {
    'feature_scaler': global_feature_scaler,
    'target_scaler': global_target_scaler,
    'client_scalers': client_scalers,
    'client_stats': client_stats
}
try:
    with open(ARTIFACTS_SAVE_PATH, 'wb') as f:
        joblib.dump(artifacts, f)
    print(f"ğŸ’¾ ç¸®æ”¾å™¨å’Œçµ±è¨ˆè³‡è¨Šå·²ä¿å­˜è‡³ '{ARTIFACTS_SAVE_PATH}'")
except Exception as e:
    print(f"âš ï¸ ä¿å­˜ artifacts æ™‚ç™¼ç”ŸéŒ¯èª¤: {e}")

# ===== æœ€çµ‚ç¢ºèª =====
print("\nğŸ“‹ ä½¿ç”¨çš„ç‰¹å¾µ:")
for i, feature in enumerate(feature_columns, 1):
    print(f"   {i:2d}. {feature}")
print("\nâœ… å…¨åŸŸè®Šæ•¸è¨­å®šå®Œæˆ:")
print(f"   - df_scaled: {df_scaled.shape}")
print(f"   - feature_columns: {len(feature_columns)} å€‹ç‰¹å¾µ")
print(f"   - target_column: {target_column}")
print(f"   - artifacts: åŒ…å«æ‰€æœ‰ç¸®æ”¾å™¨è³‡è¨Š")
```

```python
# @title ä¿®æ­£ç‰ˆ Cell 6: ä½¿ç”¨è‡ªå®šç¾© DP å¯¦ä½œï¼ˆç›¸å®¹ TF 2.14ï¼‰
import tensorflow as tf
import tensorflow_federated as tff
import numpy as np
from typing import Callable, Tuple

# ç¢ºä¿å¿…è¦è®Šæ•¸å­˜åœ¨
try:
    train_data
    CLIENT_LEARNING_RATE
    SERVER_LEARNING_RATE
    INPUT_SHAPE
    LOCAL_EPOCHS
    DP_L2_NORM_CLIP
    DP_NOISE_MULTIPLIER
except NameError:
    print("âŒ è«‹å…ˆåŸ·è¡Œå‰é¢çš„ Cells")
    raise

def create_keras_model():
    """å»ºç«‹ Keras åºè²«æ¨¡å‹ (åŠ å…¥ L2 æ­£å‰‡åŒ–)"""
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(64, activation='relu', input_shape=INPUT_SHAPE),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(
            32,
            activation='relu',
            kernel_regularizer=tf.keras.regularizers.l2(1e-4)
        ),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(
            16,
            activation='relu',
            kernel_regularizer=tf.keras.regularizers.l2(1e-4)
        ),
        tf.keras.layers.Dense(1)
    ])
    return model

print("ğŸ—ï¸ å»ºç«‹å…·æœ‰è‡ªå®šç¾©å·®åˆ†éš±ç§çš„è¯é‚¦å­¸ç¿’æ¨¡å‹èˆ‡æµç¨‹...")
print("="*60)

# è‡ªå®šç¾©å·®åˆ†éš±ç§æ¢¯åº¦è™•ç†å‡½æ•¸
@tf.function
def clip_and_add_noise_to_gradients(
    gradients: list,
    l2_norm_clip: float,
    noise_multiplier: float,
    num_microbatches: int
) -> Tuple[list, tf.Tensor]:
    """
    è£å‰ªæ¢¯åº¦ä¸¦æ·»åŠ é«˜æ–¯å™ªéŸ³

    Args:
        gradients: æ¢¯åº¦åˆ—è¡¨
        l2_norm_clip: L2 norm è£å‰ªé–¾å€¼
        noise_multiplier: å™ªéŸ³ä¹˜æ•¸
        num_microbatches: å¾®æ‰¹æ¬¡æ•¸é‡

    Returns:
        è™•ç†å¾Œçš„æ¢¯åº¦å’Œæ¢¯åº¦ç¯„æ•¸
    """
    # è¨ˆç®—å…¨åŸŸæ¢¯åº¦ç¯„æ•¸
    global_norm = tf.linalg.global_norm(gradients)

    # è£å‰ªæ¢¯åº¦
    clipped_gradients, _ = tf.clip_by_global_norm(
        gradients,
        l2_norm_clip
    )

    # æ·»åŠ é«˜æ–¯å™ªéŸ³
    noise_stddev = l2_norm_clip * noise_multiplier / np.sqrt(num_microbatches)
    noisy_gradients = []

    for grad in clipped_gradients:
        if grad is not None:
            noise = tf.random.normal(
                shape=tf.shape(grad),
                mean=0.0,
                stddev=noise_stddev,
                dtype=grad.dtype
            )
            noisy_grad = grad + noise
            noisy_gradients.append(noisy_grad)
        else:
            noisy_gradients.append(grad)

    return noisy_gradients, global_norm

# å‰µå»ºå¸¶æœ‰ DP çš„è¨“ç·´æ­¥é©Ÿ
def create_dp_train_step(model, loss_fn, optimizer, l2_norm_clip, noise_multiplier):
    """å‰µå»ºå·®åˆ†éš±ç§è¨“ç·´æ­¥é©Ÿ"""

    @tf.function
    def train_step(x, y):
        with tf.GradientTape() as tape:
            predictions = model(x, training=True)
            loss = loss_fn(y, predictions)

        # è¨ˆç®—æ¢¯åº¦
        gradients = tape.gradient(loss, model.trainable_variables)

        # æ‡‰ç”¨å·®åˆ†éš±ç§
        dp_gradients, grad_norm = clip_and_add_noise_to_gradients(
            gradients,
            l2_norm_clip,
            noise_multiplier,
            num_microbatches=tf.shape(x)[0]  # ä½¿ç”¨æ‰¹æ¬¡å¤§å°ä½œç‚ºå¾®æ‰¹æ¬¡æ•¸
        )

        # æ›´æ–°æ¬Šé‡
        optimizer.apply_gradients(zip(dp_gradients, model.trainable_variables))

        return loss, grad_norm

    return train_step

# å‰µå»ºè‡ªå®šç¾©çš„è¯é‚¦å­¸ç¿’å®¢æˆ¶ç«¯å‡½æ•¸
def create_client_fn():
    """å‰µå»ºå®¢æˆ¶ç«¯æ›´æ–°å‡½æ•¸"""

    def client_update(model_weights, dataset):
        """åŸ·è¡Œæœ¬åœ°è¨“ç·´ä¸¦è¿”å›æ›´æ–°å¾Œçš„æ¬Šé‡"""
        # å‰µå»ºæœ¬åœ°æ¨¡å‹
        model = create_keras_model()
        model.set_weights(model_weights)

        # å‰µå»ºå„ªåŒ–å™¨å’Œæå¤±å‡½æ•¸
        optimizer = tf.keras.optimizers.Adam(learning_rate=CLIENT_LEARNING_RATE)
        loss_fn = tf.keras.losses.MeanSquaredError()

        # å‰µå»º DP è¨“ç·´æ­¥é©Ÿ
        train_step = create_dp_train_step(
            model, loss_fn, optimizer,
            DP_L2_NORM_CLIP, DP_NOISE_MULTIPLIER
        )

        # è¨“ç·´æŒ‡æ¨™
        num_examples = 0
        total_loss = 0.0

        # æœ¬åœ°è¨“ç·´
        for _ in range(LOCAL_EPOCHS):
            for batch in dataset:
                x, y = batch
                loss, _ = train_step(x, y)

                batch_size = tf.shape(x)[0]
                num_examples += batch_size
                total_loss += loss * tf.cast(batch_size, tf.float32)

        # è¨ˆç®—å¹³å‡æå¤±
        average_loss = total_loss / tf.cast(num_examples, tf.float32)

        # è¿”å›æ›´æ–°å¾Œçš„æ¬Šé‡å’ŒæŒ‡æ¨™
        return model.get_weights(), num_examples, average_loss

    return client_update

# ä½¿ç”¨æ¨™æº– TFF APIï¼ˆä½†åŒ…å«æˆ‘å€‘çš„ DP å¯¦ä½œï¼‰
def model_fn():
    keras_model = create_keras_model()
    return tff.learning.models.from_keras_model(
        keras_model,
        input_spec=train_data[0].element_spec,
        loss=tf.keras.losses.MeanSquaredError(),
        metrics=[
            tf.keras.metrics.MeanAbsoluteError(),
            tf.keras.metrics.RootMeanSquaredError()
        ]
    )

# å»ºç«‹è¯é‚¦å¹³å‡æ¼”ç®—æ³•ï¼ˆä½¿ç”¨æ¨™æº–å„ªåŒ–å™¨ï¼ŒDP åœ¨å®¢æˆ¶ç«¯è™•ç†ï¼‰
try:
    federated_averaging_process = tff.learning.algorithms.build_weighted_fed_avg(
        model_fn,
        client_optimizer_fn=lambda: tf.keras.optimizers.Adam(learning_rate=CLIENT_LEARNING_RATE),
        server_optimizer_fn=lambda: tf.keras.optimizers.Adam(learning_rate=SERVER_LEARNING_RATE)
    )
    print("âœ… è¯é‚¦å­¸ç¿’æ¼”ç®—æ³•å»ºç«‹æˆåŠŸï¼")
except Exception as e:
    print(f"âŒ å»ºç«‹è¯é‚¦å­¸ç¿’æ¼”ç®—æ³•æ™‚ç™¼ç”ŸéŒ¯èª¤: {e}")
    raise

print(f"\nğŸ“Š å·®åˆ†éš±ç§è¨­å®šï¼ˆè‡ªå®šç¾©å¯¦ä½œï¼‰:")
print(f"   - L2 Norm Clip: {DP_L2_NORM_CLIP}")
print(f"   - Noise Multiplier: {DP_NOISE_MULTIPLIER}")
print(f"   - å®¢æˆ¶ç«¯å­¸ç¿’ç‡: {CLIENT_LEARNING_RATE}")
print(f"   - ä¼ºæœå™¨å­¸ç¿’ç‡: {SERVER_LEARNING_RATE}")
print(f"   - æœ¬åœ°è¨“ç·´è¼ªæ•¸: {LOCAL_EPOCHS}")

# è¨ˆç®—ç†è«–éš±ç§ä¿è­‰
def compute_privacy_guarantee(noise_multiplier, num_samples, batch_size, epochs, delta=1e-5):
    """ç°¡åŒ–çš„éš±ç§é ç®—è¨ˆç®—"""
    # æ¡æ¨£ç‡
    q = batch_size / num_samples

    # è¨“ç·´æ­¥æ•¸
    steps = epochs * num_samples / batch_size

    # ä½¿ç”¨é«˜æ–¯æ©Ÿåˆ¶çš„ç°¡åŒ– epsilon è¨ˆç®—
    # é€™æ˜¯ä¸€å€‹ä¿å®ˆçš„ä¸Šç•Œä¼°è¨ˆ
    epsilon = 2 * q * np.sqrt(steps * np.log(1/delta)) / noise_multiplier

    return epsilon

# è¨ˆç®—æ¯è¼ªçš„éš±ç§æ¶ˆè€—
samples_per_client = int(SAMPLES_PER_CLIENT * TRAIN_TEST_SPLIT_RATIO)
epsilon_per_round = compute_privacy_guarantee(
    DP_NOISE_MULTIPLIER,
    samples_per_client,
    64,  # batch size
    LOCAL_EPOCHS,
    delta=1e-5
)

print(f"\nğŸ“Š é ä¼°éš±ç§ä¿è­‰:")
print(f"   - æ¯è¼ª Îµ â‰ˆ {epsilon_per_round:.2f}")
print(f"   - é è¨ˆç¸½ Îµ (30è¼ª) â‰ˆ {epsilon_per_round * 30:.2f}")
print("   è¨»ï¼šé€™æ˜¯ä¿å®ˆä¼°è¨ˆï¼Œå¯¦éš›éš±ç§ä¿è­‰å¯èƒ½æ›´å¥½")

# é¡¯ç¤ºæ¨¡å‹çµæ§‹
model_for_summary = create_keras_model()
model_for_summary.summary()

# åˆå§‹åŒ–ä¼ºæœå™¨ç‹€æ…‹
server_state = federated_averaging_process.initialize()
print("\nâœ… ä¼ºæœå™¨ç‹€æ…‹åˆå§‹åŒ–å®Œæˆ")

# å„²å­˜ DP åƒæ•¸ä¾›å¾ŒçºŒä½¿ç”¨
dp_params = {
    'l2_norm_clip': DP_L2_NORM_CLIP,
    'noise_multiplier': DP_NOISE_MULTIPLIER,
    'epsilon_per_round': epsilon_per_round,
    'delta': 1e-5
}
```

```python
# @title ä¿®æ­£ç‰ˆ Cell 7: åŸ·è¡Œè¯é‚¦å­¸ç¿’è¨“ç·´ï¼ˆä½¿ç”¨è‡ªå®šç¾© DPï¼Œç›¸å®¹ TF 2.14ï¼‰
from tqdm.auto import tqdm
import pandas as pd
import numpy as np
import tensorflow as tf

# ç¢ºä¿å¿…è¦è®Šæ•¸å­˜åœ¨
try:
    federated_averaging_process
    server_state
    train_data
    federated_test_data
    create_keras_model
    NUM_ROUNDS
    USE_TQDM
    LOCAL_EPOCHS
    DP_NOISE_MULTIPLIER
    dp_params
    clip_and_add_noise_to_gradients
    DP_L2_NORM_CLIP
except NameError:
    print("âŒ è«‹å…ˆåŸ·è¡Œå‰é¢çš„ Cells")
    raise

print("ğŸ¯ é–‹å§‹å…·æœ‰å·®åˆ†éš±ç§çš„è¯é‚¦å­¸ç¿’è¨“ç·´...")
print("="*60)

# å‰µå»ºå…·æœ‰ DP çš„å®¢æˆ¶ç«¯æ›´æ–°å‡½æ•¸
@tf.function
def client_update_with_dp(model, dataset, optimizer, num_epochs):
    """åœ¨å®¢æˆ¶ç«¯åŸ·è¡Œå·®åˆ†éš±ç§è¨“ç·´"""
    loss_fn = tf.keras.losses.MeanSquaredError()
    mae_metric = tf.keras.metrics.MeanAbsoluteError()
    rmse_metric = tf.keras.metrics.RootMeanSquaredError()

    for _ in range(num_epochs):
        for x, y in dataset:
            with tf.GradientTape() as tape:
                predictions = model(x, training=True)
                loss = loss_fn(y, predictions)

            # è¨ˆç®—æ¢¯åº¦
            gradients = tape.gradient(loss, model.trainable_variables)

            # æ‡‰ç”¨å·®åˆ†éš±ç§ï¼šè£å‰ªä¸¦æ·»åŠ å™ªéŸ³
            dp_gradients, _ = clip_and_add_noise_to_gradients(
                gradients,
                DP_L2_NORM_CLIP,
                DP_NOISE_MULTIPLIER,
                num_microbatches=tf.shape(x)[0]
            )

            # æ›´æ–°æ¨¡å‹
            optimizer.apply_gradients(zip(dp_gradients, model.trainable_variables))

            # æ›´æ–°æŒ‡æ¨™
            mae_metric.update_state(y, predictions)
            rmse_metric.update_state(y, predictions)

    return loss, mae_metric.result(), rmse_metric.result()

# Early Stopping åƒæ•¸
best_rmse = 1e9
wait = 0
patience = 5
best_server_state = None

# éš±ç§é ç®—è¿½è¹¤
max_epsilon = DP_MAX_EPSILON if 'DP_MAX_EPSILON' in globals() else 10.0
current_epsilon = 0.0

history = []
privacy_history = []
progress_bar = tqdm(range(NUM_ROUNDS), desc="Federated Training with DP") if USE_TQDM else range(NUM_ROUNDS)

try:
    for round_num in progress_bar:
        # åŸ·è¡Œä¸€è¼ªè¯é‚¦å­¸ç¿’
        result = federated_averaging_process.next(server_state, train_data)
        server_state = result.state
        train_metrics = result.metrics['client_work']['train']

        # ç´¯ç©éš±ç§é ç®—ï¼ˆç°¡åŒ–è¨ˆç®—ï¼‰
        current_epsilon += dp_params['epsilon_per_round']

        # åœ¨å…¨åŸŸæ¸¬è©¦é›†ä¸Šè©•ä¼°æ¨¡å‹
        model_weights = federated_averaging_process.get_model_weights(server_state)

        # å»ºç«‹è©•ä¼°æ¨¡å‹
        eval_model = create_keras_model()
        eval_model.set_weights(model_weights.trainable)
        eval_model.compile(
            loss=tf.keras.losses.MeanSquaredError(),
            metrics=[tf.keras.metrics.MeanAbsoluteError(), tf.keras.metrics.RootMeanSquaredError()]
        )

        test_metrics = eval_model.evaluate(federated_test_data, verbose=0)

        # è¨˜éŒ„æŒ‡æ¨™
        round_history = {
            'round': round_num + 1,
            'train_loss': float(train_metrics['loss']),
            'train_mae': float(train_metrics['mean_absolute_error']),
            'train_rmse': float(train_metrics['root_mean_squared_error']),
            'test_loss': float(test_metrics[0]),
            'test_mae': float(test_metrics[1]),
            'test_rmse': float(test_metrics[2]),
            'epsilon': float(current_epsilon),
            'delta': dp_params['delta']
        }
        history.append(round_history)
        privacy_history.append({
            'round': round_num + 1,
            'epsilon': float(current_epsilon)
        })

        # æª¢æŸ¥æ˜¯å¦è¶…ééš±ç§é ç®—
        if current_epsilon > max_epsilon:
            print(f"\nâš ï¸ éš±ç§é ç®—å·²è€—ç›¡ï¼Îµ = {current_epsilon:.2f} > {max_epsilon}")
            print("   åœæ­¢è¨“ç·´ä»¥ä¿è­·éš±ç§ã€‚")
            break

        # Early Stopping æª¢æŸ¥
        current_rmse = test_metrics[2]
        if current_rmse < best_rmse:
            best_rmse = current_rmse
            best_server_state = server_state
            wait = 0
            print(f"\nğŸ’¾ Round {round_num + 1}: æ–°çš„æœ€ä½³ RMSE = {best_rmse:.4f}, Îµ = {current_epsilon:.2f}")
        else:
            wait += 1
            if wait >= patience:
                print(f"\nâš ï¸ Early Stopping: {patience} è¼ªæœªæ”¹å–„ï¼Œåœæ­¢è¨“ç·´ã€‚")
                print(f"   æœ€ä½³ RMSE = {best_rmse:.4f}")
                print(f"   æœ€çµ‚éš±ç§æ¶ˆè€—: Îµ = {current_epsilon:.2f}, Î´ = {dp_params['delta']}")
                break

        # æ›´æ–°é€²åº¦æ¢
        if USE_TQDM:
            progress_bar.set_postfix(
                loss=round_history['test_loss'],
                mae=round_history['test_mae'],
                rmse=round_history['test_rmse'],
                epsilon=f"{current_epsilon:.2f}",
                wait=f"{wait}/{patience}"
            )

        # å®šæœŸè¼¸å‡ºè©³ç´°è³‡è¨Š
        if (round_num + 1) % 5 == 0 or (round_num + 1) == 1:
            print(f"\n--- Round {round_num + 1}/{NUM_ROUNDS} ---")
            print(f"  Training: Loss={train_metrics['loss']:.4f}, MAE={train_metrics['mean_absolute_error']:.4f}")
            print(f"  Test:     Loss={test_metrics[0]:.4f}, MAE={test_metrics[1]:.4f}, RMSE={test_metrics[2]:.4f}")
            print(f"  Privacy:  Îµ={current_epsilon:.2f}, Î´={dp_params['delta']:.2e}")

except Exception as e:
    print(f"\nâŒ è¨“ç·´éç¨‹ä¸­ç™¼ç”ŸéŒ¯èª¤: {e}")
    print(f"   å·²å®Œæˆ {len(history)} è¼ªè¨“ç·´")
    import traceback
    traceback.print_exc()

# ä½¿ç”¨æœ€ä½³æ¨¡å‹
if best_server_state is not None:
    server_state = best_server_state
    print(f"\nâœ… è¯é‚¦å­¸ç¿’è¨“ç·´å®Œæˆï¼ä½¿ç”¨æœ€ä½³æ¨¡å‹ (RMSE={best_rmse:.4f})")
else:
    print("\nâœ… è¯é‚¦å­¸ç¿’è¨“ç·´å®Œæˆï¼")

# è½‰æ›ç‚º DataFrame
history_df = pd.DataFrame(history)
privacy_df = pd.DataFrame(privacy_history)

# æ›´æ–° dp_params
dp_params['final_epsilon'] = current_epsilon
dp_params['total_rounds'] = len(history_df)
dp_params['max_epsilon'] = max_epsilon

print(f"\nğŸ“Š è¨“ç·´çµ±è¨ˆ:")
print(f"   - ç¸½è¨“ç·´è¼ªæ•¸: {len(history_df)}")
print(f"   - æœ€çµ‚ Test RMSE: {history_df.iloc[-1]['test_rmse']:.4f}")
print(f"   - æœ€ä½³ Test RMSE: {history_df['test_rmse'].min():.4f}")
print(f"\nğŸ” éš±ç§çµ±è¨ˆ:")
print(f"   - æœ€çµ‚éš±ç§æ¶ˆè€—: Îµ = {current_epsilon:.2f}, Î´ = {dp_params['delta']:.2e}")
print(f"   - å¹³å‡æ¯è¼ª Îµ: {current_epsilon / len(history_df):.3f}")

print("\nâœ… å…¨åŸŸè®Šæ•¸æ›´æ–°:")
print("   - history_df: åŒ…å«éš±ç§é ç®—çš„è¨“ç·´æ­·å²")
print("   - privacy_df: éš±ç§æ¶ˆè€—è¿½è¹¤è¨˜éŒ„")
print("   - dp_params: æ›´æ–°å¾Œçš„å·®åˆ†éš±ç§åƒæ•¸")
```

```python
# @title Cell 10: ä¸­å¿ƒåŒ–å­¸ç¿’ vs. è¯é‚¦å­¸ç¿’ (å·²ä¿®æ­£ NotImplementError)
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# --- ç¢ºä¿å¿…è¦è®Šæ•¸å­˜åœ¨ ---
try:
    df
    feature_columns
    target_column
    SAMPLES_PER_CLIENT
    NUM_TOTAL_CLIENTS
    TRAIN_TEST_SPLIT_RATIO
    RANDOM_STATE
    CLIENT_LEARNING_RATE
    create_keras_model
    client_results_df
    history_df
    DP_L2_NORM_CLIP
    DP_NOISE_MULTIPLIER
    dp_params
except NameError:
    print("âŒ è«‹å…ˆåŸ·è¡Œå‰é¢çš„ Cells")
    raise

print("ğŸ”¬ é€²éšåˆ†æï¼šè¯é‚¦å­¸ç¿’(with DP) vs. ä¸­å¿ƒåŒ–å­¸ç¿’(with DP) vs. ä¸­å¿ƒåŒ–å­¸ç¿’(without DP)")
print("="*60)

# --- 1. æº–å‚™ä¸­å¿ƒåŒ–å­¸ç¿’çš„è³‡æ–™é›† ---
centralized_df = df.sample(n=min(len(df), SAMPLES_PER_CLIENT * NUM_TOTAL_CLIENTS), random_state=RANDOM_STATE)
print(f"ğŸ“Š ç‚ºä¸­å¿ƒåŒ–å­¸ç¿’æŠ½æ¨£äº† {len(centralized_df):,} ç­†è¨˜éŒ„ã€‚")

X_cen = centralized_df[feature_columns]
y_cen = centralized_df[[target_column]]

# åˆ†å‰²è¨“ç·´é›†èˆ‡æ¸¬è©¦é›†
X_train_cen, X_test_cen, y_train_cen, y_test_cen = train_test_split(
    X_cen, y_cen, test_size=(1 - TRAIN_TEST_SPLIT_RATIO), random_state=RANDOM_STATE
)

# å‰µå»ºä¸­å¿ƒåŒ–çš„ç¸®æ”¾å™¨
centralized_feature_scaler = StandardScaler()
centralized_target_scaler = StandardScaler()

# æ“¬åˆä¸¦è½‰æ›è¨“ç·´è³‡æ–™
X_train_cen_scaled = centralized_feature_scaler.fit_transform(X_train_cen)
y_train_cen_scaled = centralized_target_scaler.fit_transform(y_train_cen)

# è½‰æ›æ¸¬è©¦è³‡æ–™
X_test_cen_scaled = centralized_feature_scaler.transform(X_test_cen)
y_test_cen_scaled = centralized_target_scaler.transform(y_test_cen)

print(f"   - è¨“ç·´é›†å¤§å°: {len(X_train_cen):,}")
print(f"   - æ¸¬è©¦é›†å¤§å°: {len(X_test_cen):,}")

# --- 2. è¨“ç·´ä¸­å¿ƒåŒ–æ¨¡å‹ï¼ˆç„¡DPï¼‰ ---
print("\nğŸ’ª è¨“ç·´ä¸­å¿ƒåŒ–æ¨¡å‹ï¼ˆç„¡å·®åˆ†éš±ç§ï¼‰...")
centralized_model_no_dp = create_keras_model()

centralized_model_no_dp.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=CLIENT_LEARNING_RATE),
    loss='mean_squared_error',
    metrics=['mean_absolute_error', tf.keras.metrics.RootMeanSquaredError()]
)

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_root_mean_squared_error',
    patience=5,
    restore_best_weights=True,
    mode='min'
)

history_cen_no_dp = centralized_model_no_dp.fit(
    X_train_cen_scaled, y_train_cen_scaled,
    epochs=50,
    batch_size=64,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=0
)

print(f"âœ… ä¸­å¿ƒåŒ–æ¨¡å‹ï¼ˆç„¡DPï¼‰è¨“ç·´å®Œæˆï¼è¨“ç·´äº† {len(history_cen_no_dp.history['loss'])} å€‹ epoch")


# --- (ä¿®æ­£éƒ¨åˆ†) è¼”åŠ©å‡½å¼ï¼šä¿®æ­£ np.sqrt -> tf.sqrt ---
# ç‚ºäº†è®“æ­¤ Cell ç¨ç«‹ï¼Œå°‡ä¿®æ­£å¾Œçš„å‡½å¼æ”¾åœ¨é€™è£¡
@tf.function
def clip_and_add_noise_to_gradients_tf(
    gradients: list,
    l2_norm_clip: float,
    noise_multiplier: float,
    num_microbatches: tf.Tensor
):
    """è£å‰ªæ¢¯åº¦ä¸¦æ·»åŠ é«˜æ–¯å™ªéŸ³ (TensorFlow ç‰ˆæœ¬)"""
    # è¨ˆç®—å…¨åŸŸæ¢¯åº¦ç¯„æ•¸
    global_norm = tf.linalg.global_norm(gradients)

    # è£å‰ªæ¢¯åº¦
    clipped_gradients, _ = tf.clip_by_global_norm(gradients, l2_norm_clip)

    # *** ä¿®æ­£é» ***
    # ä½¿ç”¨ tf.sqrt ä¸¦å°‡ num_microbatches è½‰æ›ç‚º float32
    noise_stddev = l2_norm_clip * noise_multiplier / tf.sqrt(tf.cast(num_microbatches, tf.float32))

    # æ·»åŠ é«˜æ–¯å™ªéŸ³
    noisy_gradients = []
    for grad in clipped_gradients:
        if grad is not None:
            noise = tf.random.normal(
                shape=tf.shape(grad),
                mean=0.0,
                stddev=noise_stddev,
                dtype=grad.dtype
            )
            noisy_grad = grad + noise
            noisy_gradients.append(noisy_grad)
        else:
            noisy_gradients.append(grad)

    return noisy_gradients, global_norm

# --- 3. è¨“ç·´ä¸­å¿ƒåŒ–æ¨¡å‹ï¼ˆæœ‰DPï¼‰ ---
print("\nğŸ’ª è¨“ç·´ä¸­å¿ƒåŒ–æ¨¡å‹ï¼ˆæœ‰å·®åˆ†éš±ç§ï¼‰...")

# å‰µå»ºè‡ªå®šç¾©è¨“ç·´å¾ªç’°ä»¥åŠ å…¥DP
centralized_model_dp = create_keras_model()
optimizer_dp = tf.keras.optimizers.Adam(learning_rate=CLIENT_LEARNING_RATE)

# è½‰æ›ç‚º TensorFlow Dataset
train_dataset_cen = tf.data.Dataset.from_tensor_slices((X_train_cen_scaled, y_train_cen_scaled))
train_dataset_cen = train_dataset_cen.batch(64).shuffle(1000)

# è¨“ç·´å¾ªç’°
@tf.function
def train_step_dp(x, y, model, optimizer):
    with tf.GradientTape() as tape:
        predictions = model(x, training=True)
        loss = tf.keras.losses.mean_squared_error(y, predictions)
        loss = tf.reduce_mean(loss)

    # è¨ˆç®—æ¢¯åº¦
    gradients = tape.gradient(loss, model.trainable_variables)

    # æ‡‰ç”¨å·®åˆ†éš±ç§ï¼ˆä½¿ç”¨ä¿®æ­£å¾Œçš„ TensorFlow ç‰ˆæœ¬å‡½æ•¸ï¼‰
    dp_gradients, _ = clip_and_add_noise_to_gradients_tf(
        gradients,
        DP_L2_NORM_CLIP,
        DP_NOISE_MULTIPLIER,
        num_microbatches=tf.shape(x)[0]
    )
    # æ›´æ–°æ¬Šé‡
    optimizer.apply_gradients(zip(dp_gradients, model.trainable_variables))
    return loss

# è¨“ç·´æ¨¡å‹
history_cen_dp = {'loss': [], 'val_loss': []}
best_val_loss = float('inf')
patience_counter = 0

for epoch in range(50):
    # è¨“ç·´
    epoch_loss = []
    for x_batch, y_batch in train_dataset_cen:
        loss = train_step_dp(x_batch, y_batch, centralized_model_dp, optimizer_dp)
        epoch_loss.append(loss)

    # é©—è­‰
    val_predictions = centralized_model_dp(X_test_cen_scaled, training=False)
    val_loss = tf.reduce_mean(tf.keras.losses.mean_squared_error(y_test_cen_scaled, val_predictions))

    history_cen_dp['loss'].append(np.mean(epoch_loss))
    history_cen_dp['val_loss'].append(val_loss.numpy())

    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 5:
            break

print(f"âœ… ä¸­å¿ƒåŒ–æ¨¡å‹ï¼ˆæœ‰DPï¼‰è¨“ç·´å®Œæˆï¼è¨“ç·´äº† {len(history_cen_dp['loss'])} å€‹ epoch")

# --- 4. è©•ä¼°æ‰€æœ‰æ¨¡å‹ ---
# è¯é‚¦å­¸ç¿’ï¼ˆæœ‰DPï¼‰
mae_fl = client_results_df['mae'].mean()
rmse_fl = client_results_df['rmse'].mean()

# ä¸­å¿ƒåŒ–å­¸ç¿’ï¼ˆç„¡DPï¼‰
scaled_preds_no_dp = centralized_model_no_dp.predict(X_test_cen_scaled, verbose=0)
unscaled_preds_no_dp = centralized_target_scaler.inverse_transform(scaled_preds_no_dp)
unscaled_true_cen = centralized_target_scaler.inverse_transform(y_test_cen_scaled.reshape(-1, 1))
mae_cen_no_dp = np.mean(np.abs(unscaled_true_cen - unscaled_preds_no_dp))
rmse_cen_no_dp = np.sqrt(np.mean((unscaled_true_cen - unscaled_preds_no_dp)**2))

# ä¸­å¿ƒåŒ–å­¸ç¿’ï¼ˆæœ‰DPï¼‰
scaled_preds_dp = centralized_model_dp.predict(X_test_cen_scaled, verbose=0)
unscaled_preds_dp = centralized_target_scaler.inverse_transform(scaled_preds_dp)
mae_cen_dp = np.mean(np.abs(unscaled_true_cen - unscaled_preds_dp))
rmse_cen_dp = np.sqrt(np.mean((unscaled_true_cen - unscaled_preds_dp)**2))

# --- 5. çµæœæ¯”è¼ƒèˆ‡è¦–è¦ºåŒ– ---
print("\nğŸ“Š æ¨¡å‹è¡¨ç¾æ¯”è¼ƒ (åŸºæ–¼çœŸå¯¦æ•¸å€¼):")
print(f"  - è¯é‚¦å­¸ç¿’ (with DP):    MAE = {mae_fl:.4f}, RMSE = {rmse_fl:.4f}")
print(f"  - ä¸­å¿ƒåŒ–å­¸ç¿’ (with DP):  MAE = {mae_cen_dp:.4f}, RMSE = {rmse_cen_dp:.4f}")
print(f"  - ä¸­å¿ƒåŒ–å­¸ç¿’ (no DP):    MAE = {mae_cen_no_dp:.4f}, RMSE = {rmse_cen_no_dp:.4f}")

# è¨ˆç®—éš±ç§æˆæœ¬
dp_cost_fl = ((mae_fl - mae_cen_no_dp) / mae_cen_no_dp) * 100
dp_cost_cen = ((mae_cen_dp - mae_cen_no_dp) / mae_cen_no_dp) * 100

print(f"\nğŸ“ˆ å·®åˆ†éš±ç§çš„æˆæœ¬:")
print(f"  - è¯é‚¦å­¸ç¿’ DP æˆæœ¬: {dp_cost_fl:.1f}% MAE å¢åŠ ")
print(f"  - ä¸­å¿ƒåŒ– DP æˆæœ¬: {dp_cost_cen:.1f}% MAE å¢åŠ ")

# ç¹ªè£½æ¯”è¼ƒåœ–
comparison_data = {
    'Method': ['Federated\n(with DP)', 'Centralized\n(with DP)', 'Centralized\n(no DP)',
               'Federated\n(with DP)', 'Centralized\n(with DP)', 'Centralized\n(no DP)'],
    'Metric': ['MAE', 'MAE', 'MAE', 'RMSE', 'RMSE', 'RMSE'],
    'Value': [mae_fl, mae_cen_dp, mae_cen_no_dp, rmse_fl, rmse_cen_dp, rmse_cen_no_dp]
}
comparison_df = pd.DataFrame(comparison_data)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.set_style("whitegrid")
fig.suptitle('Comprehensive Comparison: Federated vs Centralized Learning with/without DP', fontsize=16)

# 1. MAE æ¯”è¼ƒ
mae_data = comparison_df[comparison_df['Metric'] == 'MAE']
bars1 = axes[0].bar(range(3), mae_data['Value'],
                    color=['royalblue', 'orange', 'green'], alpha=0.7)
axes[0].set_title('Mean Absolute Error (MAE) Comparison')
axes[0].set_ylabel('MAE (Lower is Better)')
axes[0].set_xticks(range(3))
axes[0].set_xticklabels(['Federated\n(with DP)', 'Centralized\n(with DP)', 'Centralized\n(no DP)'])
axes[0].bar_label(bars1, fmt='%.4f')

# 2. RMSE æ¯”è¼ƒ
rmse_data = comparison_df[comparison_df['Metric'] == 'RMSE']
bars2 = axes[1].bar(range(3), rmse_data['Value'],
                    color=['royalblue', 'orange', 'green'], alpha=0.7)
axes[1].set_title('Root Mean Square Error (RMSE) Comparison')
axes[1].set_ylabel('RMSE (Lower is Better)')
axes[1].set_xticks(range(3))
axes[1].set_xticklabels(['Federated\n(with DP)', 'Centralized\n(with DP)', 'Centralized\n(no DP)'])
axes[1].bar_label(bars2, fmt='%.4f')

# 3. éš±ç§-æ•ˆç”¨æ¬Šè¡¡
privacy_levels = [dp_params['final_epsilon'], dp_params['final_epsilon'], 0]  # 0 è¡¨ç¤ºç„¡DP
utility_loss = [dp_cost_fl, dp_cost_cen, 0]
colors = ['royalblue', 'orange', 'green']
markers = ['o', 's', '^']

for i, (method, eps, loss) in enumerate(zip(['Federated (DP)', 'Centralized (DP)', 'Centralized (No DP)'],
                                            privacy_levels, utility_loss)):
    axes[2].scatter(eps, loss, color=colors[i], s=200, marker=markers[i],
                   label=method, edgecolors='black', linewidth=2)

axes[2].set_title('Privacy-Utility Tradeoff')
axes[2].set_xlabel('Privacy Cost (Îµ) - Lower is More Private')
axes[2].set_ylabel('Utility Loss (% MAE Increase)')
axes[2].legend()
axes[2].grid(True, alpha=0.3)
axes[2].set_xlim(-0.5, max(privacy_levels) + 1)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# è¨“ç·´æ­·å²æ¯”è¼ƒ
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
if hasattr(history_cen_no_dp, 'history'):
    ax.plot(history_cen_no_dp.history['val_loss'], label='Centralized (No DP) Val Loss', marker='o')
ax.plot(history_cen_dp['val_loss'], label='Centralized (With DP) Val Loss', marker='s')
ax.plot(history_df['test_loss'], label='Federated (With DP) Test Loss', marker='^')
ax.set_title('Training Progress Comparison')
ax.set_xlabel('Epoch/Round')
ax.set_ylabel('Loss (Scaled)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()

print("\nğŸ“Š åˆ†æç¸½çµ:")
print("1. å·®åˆ†éš±ç§å½±éŸ¿:")
print(f"   - ä¸­å¿ƒåŒ–å­¸ç¿’åŠ å…¥DPå¾Œï¼ŒMAEå¢åŠ äº† {dp_cost_cen:.1f}%")
print(f"   - è¯é‚¦å­¸ç¿’(å·²å«DP)ç›¸æ¯”ç„¡DPä¸­å¿ƒåŒ–ï¼ŒMAEå¢åŠ äº† {dp_cost_fl:.1f}%")
print("\n2. è¯é‚¦å­¸ç¿’å„ªå‹¢:")
print("   - æä¾›å·®åˆ†éš±ç§ä¿è­·çš„åŒæ™‚ï¼Œä¿æŒäº†ç«¶çˆ­æ€§çš„æ€§èƒ½")
print("   - è³‡æ–™å¾æœªé›¢é–‹æœ¬åœ°ï¼Œæä¾›é¡å¤–çš„éš±ç§ä¿è­·å±¤")
print("   - å„å®¢æˆ¶ç«¯ç¨ç«‹ç¸®æ”¾é€²ä¸€æ­¥å¢å¼·éš±ç§")
```

```python
# ===== Cell 11: éš±ç§ä¿è­·èˆ‡é€šè¨Šæˆæœ¬åˆ†æ (ç¢ºä¿è®Šæ•¸æ­£ç¢º) =====
# @title Cell 11: éš±ç§ä¿è­·èˆ‡é€šè¨Šæˆæœ¬åˆ†æ (æ›´æ–°ç‰ˆ)
import numpy as np

# ç¢ºä¿å¿…è¦è®Šæ•¸å­˜åœ¨
try:
    SAMPLES_PER_CLIENT
    final_model
    history_df
    NUM_ROUNDS
    NUM_TOTAL_CLIENTS
    feature_columns
except NameError:
    print("âŒ è«‹å…ˆåŸ·è¡Œå‰é¢çš„ Cells")
    raise

print("ğŸ” è¯é‚¦å­¸ç¿’éš±ç§ä¿è­·èˆ‡é€šè¨Šæˆæœ¬åˆ†æ")
print("="*60)

# --- 1. è³‡æ–™éš±ç§ä¿è­·å„ªå‹¢ ---
print("ğŸ“Š è³‡æ–™éš±ç§ä¿è­·å„ªå‹¢:")
print("1. åŸå§‹è³‡æ–™ä¿ç•™åœ¨æœ¬åœ°:")
print(f"   - æ¯å€‹åŸºåœ°å°å„è‡ªçš„ {SAMPLES_PER_CLIENT:,} ç­†æ•æ„Ÿç‡Ÿé‹è³‡æ–™ç„¡éœ€ä¸Šå‚³è‡³ä¸­å¤®ä¼ºæœå™¨ã€‚")
print("   - å¾æ ¹æœ¬ä¸Šæœçµ•äº†ä¸­å¤®è³‡æ–™åº«è¢«æ”»æ“Šå°è‡´å¤§è¦æ¨¡è³‡æ–™æ´©æ¼çš„é¢¨éšªã€‚")

print("\n2. åƒ…å‚³è¼¸æ¨¡å‹æ›´æ–°:")
print("   - æºé€šçš„å…§å®¹æ˜¯æ¨¡å‹çš„æ¬Šé‡æ¢¯åº¦ï¼Œè€ŒéåŸå§‹è³‡æ–™ã€‚")
print("   - é€™äº›æ•¸å€¼æœ¬èº«ä¸åŒ…å«ç›´æ¥å¯è§£è®€çš„ç”¨æˆ¶è³‡è¨Šã€‚")

print("\n3. å¯æ•´åˆå·®åˆ†éš±ç§ (Differential Privacy):")
print("   - å¯åœ¨å®¢æˆ¶ç«¯ä¸Šå‚³æ¨¡å‹æ›´æ–°å‰ï¼Œå°å…¶æ·»åŠ å°‘é‡æ•¸å­¸å™ªéŸ³ã€‚")
print("   - é€™èƒ½é€²ä¸€æ­¥æä¾›åš´æ ¼çš„æ•¸å­¸è­‰æ˜ï¼Œä¿è­·å–®ä¸€åŸºåœ°å°çš„è³‡æ–™ä¸è¢«å¾æ¨¡å‹æ›´æ–°ä¸­åæ¨å‡ºä¾†ã€‚")

print("\n4. å„å®¢æˆ¶ç«¯ç¨ç«‹ç¸®æ”¾çš„é¡å¤–ä¿è­·:")
print("   - æ¯å€‹å®¢æˆ¶ç«¯ä½¿ç”¨è‡ªå·±çš„ StandardScalerï¼Œé€²ä¸€æ­¥æ··æ·†åŸå§‹æ•¸æ“šåˆ†ä½ˆã€‚")
print("   - å³ä½¿æ”»æ“Šè€…ç²å¾—æ¨¡å‹æ›´æ–°ï¼Œä¹Ÿé›£ä»¥æ¨æ–·åŸå§‹è³‡æ–™çš„çœŸå¯¦ç¯„åœã€‚")

# --- 2. é€šè¨Šæˆæœ¬å‹•æ…‹è¨ˆç®— ---
print("\n" + "="*60)
print("ğŸ“¡ é€šè¨Šæˆæœ¬æ¯”è¼ƒ:")

# è¨ˆç®—æ¨¡å‹åƒæ•¸æ•¸é‡
num_params = final_model.count_params()
print(f"   - æ¨¡å‹ç¸½åƒæ•¸æ•¸é‡: {num_params:,}")

# å‡è¨­æ¯å€‹åƒæ•¸ç‚º 32 ä½å…ƒæµ®é»æ•¸ (4 Bytes)
bytes_per_param = 4

# è¯é‚¦å­¸ç¿’é€šè¨Šæˆæœ¬
actual_rounds = len(history_df)  # å¯¦éš›è¨“ç·´è¼ªæ•¸ï¼ˆå¯èƒ½å›  Early Stopping æå‰çµæŸï¼‰
fl_cost_per_round = num_params * bytes_per_param * NUM_TOTAL_CLIENTS  # æ‰€æœ‰å®¢æˆ¶ç«¯ä¸Šå‚³
total_fl_cost = fl_cost_per_round * actual_rounds
print(f"\nè¯é‚¦å­¸ç¿’ (å‚³è¼¸æ¨¡å‹æ›´æ–°):")
print(f"   - æ¯è¼ªæˆæœ¬: {fl_cost_per_round / 1024:.2f} KB")
print(f"   - ç¸½æˆæœ¬ ({actual_rounds} è¼ª): {total_fl_cost / (1024**2):.2f} MB")
if actual_rounds < NUM_ROUNDS:
    saved_rounds = NUM_ROUNDS - actual_rounds
    saved_cost = saved_rounds * fl_cost_per_round / (1024**2)
    print(f"   - Early Stopping ç¯€çœ: {saved_cost:.2f} MB ({saved_rounds} è¼ª)")

# ä¸­å¿ƒåŒ–å­¸ç¿’é€šè¨Šæˆæœ¬
num_features = len(feature_columns)
total_samples = SAMPLES_PER_CLIENT * NUM_TOTAL_CLIENTS
centralized_cost = total_samples * (num_features + 1) * bytes_per_param
print(f"\nä¸­å¿ƒåŒ–å­¸ç¿’ (ä¸Šå‚³æ‰€æœ‰æŠ½æ¨£è³‡æ–™):")
print(f"   - ç¸½æˆæœ¬: {centralized_cost / (1024**2):.2f} MB")

# è¨ˆç®—ç¯€çœæ¯”ä¾‹
communication_saving = (1 - (total_fl_cost / centralized_cost)) * 100
print(f"\nâœ… ä½¿ç”¨è¯é‚¦å­¸ç¿’é è¨ˆå¯ç¯€çœç´„ {communication_saving:.1f}% çš„é€šè¨Šæˆæœ¬ã€‚")

# é¡å¤–çš„éš±ç§å„ªå‹¢
print("\nğŸ›¡ï¸ é¡å¤–çš„éš±ç§ä¿è­·æªæ–½:")
print("1. **åŠ å¯†é€šè¨Š**: æ¨¡å‹æ›´æ–°å¯ä½¿ç”¨å®‰å…¨èšåˆå”è­°åŠ å¯†å‚³è¼¸")
print("2. **å®¢æˆ¶ç«¯ä¸Ÿæ£„**: æ”¯æ´å®¢æˆ¶ç«¯éš¨æ™‚é€€å‡ºï¼Œä¸å½±éŸ¿æ•´é«”è¨“ç·´")
print("3. **æ¨¡å‹å£“ç¸®**: å¯é€²ä¸€æ­¥ä½¿ç”¨é‡åŒ–æˆ–ç¨€ç–åŒ–æ¸›å°‘é€šè¨Šæˆæœ¬")
print("4. **è¯é‚¦é·ç§»å­¸ç¿’**: åªæ›´æ–°éƒ¨åˆ†å±¤ï¼Œé€²ä¸€æ­¥ä¿è­·éš±ç§")
```

```python
# ===== Cell 12: å³æ™‚ç›£æ§èˆ‡éƒ¨ç½²å»ºè­° (å®Œæ•´ç‰ˆ) =====
# @title Cell 12: å³æ™‚ç›£æ§èˆ‡éƒ¨ç½²å»ºè­° (ä½¿ç”¨å…¨åŸŸç¸®æ”¾å™¨)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ç¢ºä¿å¿…è¦è®Šæ•¸å­˜åœ¨
try:
    predict_efficiency
    final_model
    global_feature_scaler
    global_target_scaler
    df
    feature_columns
    history_df
except NameError:
    print("âŒ è«‹å…ˆåŸ·è¡Œå‰é¢çš„ Cells")
    raise

print("ğŸš€ å³æ™‚é æ¸¬èˆ‡éƒ¨ç½²å»ºè­°")
print("="*60)

# --- 1. æ¨¡æ“¬å³æ™‚ç›£æ§å ´æ™¯ ---
print("ğŸ“Š æ¨¡æ“¬ 24 å°æ™‚å³æ™‚ç›£æ§å ´æ™¯...")
monitoring_data = [
    {'time': '08:00', 'connected_users': 31, 'network_load_perc': 73.8},
    {'time': '12:00', 'connected_users': 16, 'network_load_perc': 38.1},
    {'time': '18:00', 'connected_users': 29, 'network_load_perc': 69.0},
    {'time': '22:00', 'connected_users': 14, 'network_load_perc': 33.3}
]

monitoring_results = []
default_values = df[feature_columns].mean().to_dict()

# å®šç¾©æ•ˆç‡é–¾å€¼
THRESHOLDS = {
    "Excellent": 0.8,
    "Good": 0.6,
    "Needs Improvement": 0.4
}

print("-" * 60)
for entry in monitoring_data:
    input_data = default_values.copy()
    input_data['num_ues'] = entry['connected_users']
    input_data['network_load'] = entry['network_load_perc']

    # ä½¿ç”¨å…¨åŸŸç¸®æ”¾å™¨é€²è¡Œé æ¸¬
    try:
        predicted_eff = predict_efficiency(input_data, final_model, global_feature_scaler, global_target_scaler)
    except Exception as e:
        print(f"âŒ é æ¸¬å¤±æ•—: {e}")
        continue

    entry['predicted_efficiency'] = predicted_eff

    # åˆ¤æ–·ç‹€æ…‹
    if predicted_eff >= THRESHOLDS['Excellent']:
        status = "ğŸŸ¢ Excellent"
        recommendation = "ç³»çµ±é‹ä½œè‰¯å¥½ï¼ŒæŒçºŒç›£æ§ã€‚"
    elif predicted_eff >= THRESHOLDS['Good']:
        status = "ğŸŸ¡ Good"
        recommendation = "æ•ˆç‡è‰¯å¥½ï¼Œæ³¨æ„è² è¼‰è®ŠåŒ–ã€‚"
    elif predicted_eff >= THRESHOLDS['Needs Improvement']:
        status = "ğŸŸ  Fair"
        recommendation = "æ•ˆç‡ä¸€èˆ¬ï¼Œè€ƒæ…®è¼•åº¦è³‡æºèª¿æ•´ã€‚"
    else:
        status = "ğŸ”´ Poor"
        recommendation = "æ•ˆç‡ä½ä¸‹ï¼Œå»ºè­°ç«‹å³é€²è¡Œè³‡æºå„ªåŒ–ï¼"

    entry['status'] = status
    monitoring_results.append(entry)

    print(f"æ™‚é–“: {entry['time']}")
    print(f"  - é€£ç·šç”¨æˆ¶æ•¸: {entry['connected_users']}, ç¶²è·¯è² è¼‰: {entry['network_load_perc']}%")
    print(f"  â†’ é æ¸¬æ•ˆç‡: {predicted_eff:.3f}")
    print(f"  â†’ ç‹€æ…‹: {status}")
    print(f"  â†’ å»ºè­°: {recommendation}\n")

if monitoring_results:
    monitoring_df = pd.DataFrame(monitoring_results)

    # --- 2. è¦–è¦ºåŒ–ç›£æ§çµæœ ---
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True, gridspec_kw={'height_ratios': [2, 1]})
    sns.set_style("whitegrid")
    fig.suptitle('24-Hour Resource Allocation Efficiency Monitoring', fontsize=16)

    # ä¸Šæ–¹åœ–ï¼šé æ¸¬æ•ˆç‡
    ax1.plot(monitoring_df['time'], monitoring_df['predicted_efficiency'], marker='o', linestyle='-',
             color='blue', linewidth=2, markersize=8, label='Predicted Efficiency')
    ax1.axhline(y=THRESHOLDS['Excellent'], color='green', linestyle='--',
                label=f"Excellent Threshold ({THRESHOLDS['Excellent']})")
    ax1.axhline(y=THRESHOLDS['Good'], color='orange', linestyle='--',
                label=f"Good Threshold ({THRESHOLDS['Good']})")
    ax1.axhline(y=THRESHOLDS['Needs Improvement'], color='red', linestyle='--',
                label=f"Needs Improvement ({THRESHOLDS['Needs Improvement']})")
    ax1.set_ylabel('Predicted Resource Allocation Efficiency')
    ax1.set_title('Predicted Efficiency vs. Time')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # ä¸‹æ–¹åœ–ï¼šç¶²è·¯è² è¼‰
    ax2.bar(monitoring_df['time'], monitoring_df['connected_users'], color='skyblue', label='Connected Users')
    ax2.set_xlabel('Time')
    ax2.set_ylabel('Connected Users')
    ax2.set_title('Network Load Variation')
    ax2.grid(False)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# --- 3. éƒ¨ç½²å»ºè­° ---
print("\n" + "="*60)
print("ğŸ¯ æœ€çµ‚éƒ¨ç½²èˆ‡ç¶­é‹å»ºè­°ï¼ˆé‡å°æ”¹é€²å¾Œçš„æ¶æ§‹ï¼‰")
print("="*60)
print(f"""
1. **æ”¹é€²å¾Œçš„é‚Šç·£éƒ¨ç½²æ¶æ§‹**:
   - åœ¨æ¯å€‹åŸºåœ°å°éƒ¨ç½²è¼•é‡åŒ–çš„ `final_model`ï¼Œä½¿ç”¨å…¨åŸŸåŠ æ¬Šå¹³å‡çš„ç¸®æ”¾å™¨ã€‚
   - Early Stopping ç¢ºä¿æ¨¡å‹ä¸æœƒéåº¦æ“¬åˆï¼Œæé«˜æ³›åŒ–èƒ½åŠ›ã€‚
   - ä½¿ç”¨ Adam å„ªåŒ–å™¨å’Œ L2 æ­£å‰‡åŒ–ï¼Œæä¾›æ›´ç©©å®šçš„è¨“ç·´éç¨‹ã€‚
   - å¯¦éš›è¨“ç·´åƒ…éœ€ {len(history_df)} è¼ªï¼Œç¯€çœ {((NUM_ROUNDS - len(history_df)) / NUM_ROUNDS * 100):.1f}% çš„æ™‚é–“ã€‚

2. **å„å®¢æˆ¶ç«¯ç¨ç«‹ç¸®æ”¾çš„å„ªå‹¢**:
   - **éš±ç§ä¿è­·åŠ å¼·**: æ¯å€‹å®¢æˆ¶ç«¯çš„è³‡æ–™åˆ†ä½ˆè³‡è¨Šä¸æœƒæ´©æ¼ã€‚
   - **é­¯æ£’æ€§æå‡**: å°è³‡æ–™åˆ†ä½ˆå·®ç•°è¼ƒå¤§çš„å®¢æˆ¶ç«¯æ›´å‹å¥½ã€‚
   - **æ¨è«–ä¸€è‡´æ€§**: ä½¿ç”¨å…¨åŸŸåŠ æ¬Šå¹³å‡ç¸®æ”¾å™¨ç¢ºä¿æ¨è«–æ™‚çš„ä¸€è‡´æ€§ã€‚

3. **æ¨¡å‹æ›´æ–°ç­–ç•¥ï¼ˆæ”¹é€²ç‰ˆï¼‰**:
   - **è‡ªé©æ‡‰è¨“ç·´è¼ªæ•¸**: Early Stopping è‡ªå‹•æ±ºå®šæœ€ä½³è¨“ç·´è¼ªæ•¸ã€‚
   - **æ•ˆèƒ½ç›£æ§**: æŒçºŒç›£æ§å„å®¢æˆ¶ç«¯çš„ RMSEï¼Œåµæ¸¬ç•°å¸¸ã€‚
   - **å‹•æ…‹å­¸ç¿’ç‡**: å¯è€ƒæ…®å¯¦æ–½å­¸ç¿’ç‡è¡°æ¸›ç­–ç•¥ã€‚
   - **æ¨¡å‹ç‰ˆæœ¬æ§åˆ¶**: ä¿å­˜æ¯æ¬¡è¨“ç·´çš„æœ€ä½³æ¨¡å‹å’Œè¨“ç·´æ­·å²ã€‚

4. **æ“´å±•æ€§èˆ‡ç¶­è­·**:
   - æ–°å®¢æˆ¶ç«¯åŠ å…¥æ™‚ï¼Œåˆå§‹ä½¿ç”¨å…¨åŸŸç¸®æ”¾å™¨ï¼Œç´¯ç©è¶³å¤ è³‡æ–™å¾Œå»ºç«‹è‡ªå·±çš„ç¸®æ”¾å™¨ã€‚
   - å®šæœŸæ›´æ–°å…¨åŸŸç¸®æ”¾å™¨çš„åŠ æ¬Šå¹³å‡åƒæ•¸ã€‚
   - ä¿å­˜æ¯æ¬¡è¨“ç·´çš„æœ€ä½³æ¨¡å‹å’Œå°æ‡‰çš„ Early Stopping è¼ªæ•¸ã€‚
   - æ”¯æ´å¢é‡å­¸ç¿’ï¼Œç„¡éœ€å¾é ­è¨“ç·´ã€‚

5. **æ•ˆèƒ½æŒ‡æ¨™è¿½è¹¤**:
   - è¿½è¹¤å„è¼ªçš„ Train/Test RMSE å·®ç•°ï¼Œç›£æ§éæ“¬åˆé¢¨éšªã€‚
   - è¨˜éŒ„ Early Stopping è§¸ç™¼é »ç‡ï¼Œèª¿æ•´ patience åƒæ•¸ã€‚
   - åˆ†æå„å®¢æˆ¶ç«¯çš„æ•ˆèƒ½å·®ç•°ï¼Œè­˜åˆ¥éœ€è¦ç‰¹åˆ¥é—œæ³¨çš„ç¯€é»ã€‚
   - ç›£æ§é€šè¨Šæˆæœ¬ç¯€çœæ¯”ä¾‹ã€‚

6. **å¯¦éš›éƒ¨ç½²æª¢æŸ¥æ¸…å–®**:
   âœ“ æ¨¡å‹æª”æ¡ˆ: {MODEL_SAVE_PATH}
   âœ“ ç¸®æ”¾å™¨æª”æ¡ˆ: {ARTIFACTS_SAVE_PATH}
   âœ“ æ‰¹æ¬¡å¤§å°: 64
   âœ“ å„ªåŒ–å™¨: Adam (client lr={CLIENT_LEARNING_RATE}, server lr={SERVER_LEARNING_RATE})
   âœ“ æ­£å‰‡åŒ–: L2 (Î»=1e-4)
   âœ“ Early Stopping: patience=3
""")

print("\n" + "="*80)
print("ğŸ‰ğŸ‰ğŸ‰ æ­å–œï¼æ‚¨å·²æˆåŠŸè§£æ±ºéåº¦æ“¬åˆå•é¡Œä¸¦éƒ¨ç½²æ”¹é€²çš„è¯é‚¦å­¸ç¿’æ–¹æ¡ˆï¼ ğŸ‰ğŸ‰ğŸ‰")
print("="*80)
```
